```python
import math

# 計算「所有人生日都不同」的機率
def birthday_problem(n, days=366):
    prob = 1.0
    for i in range(n):
        prob *= (days - i) / days
        print(prob)
    return prob

# 計算至少兩個人生日相同的機率
def probability_of_match(n, days=366):
    return 1 - birthday_problem(n, days)

# 找到超過 50% 的人數
people_count = 0
threshold = 0.5

for i in range(1, 100):
    if probability_of_match(i) > threshold:
        people_count = i
        break
```
